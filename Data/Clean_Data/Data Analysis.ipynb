```python
colors_dict = {
    'Abies': '#132a13',
    'Alnus': '#31572c',
    'Betula': '#4f772d',
    'Corylus': '#90a955',
    'Juniperus': '#ecf39e',
    'Larix': '#f9c80e',
    'Picea': '#99582a',
    'Pinus': '#bb9457',
    'Populus': '#8D6E63',
    'Salix': '#3E2723',
    'Sorbus': '#FF9800'
}
```

```python
genus = ForestGenus[ForestGenus['COUNTRY'] == 'Finland']
genus_counts = genus['Genus'].value_counts()
genus_percents = genus_counts / genus_counts.sum()
genus_percents_filtered = genus_percents[genus_percents > 0.03]
colors_palette = genus_percents_filtered.index.map(colors_dict.get)
plt.pie(genus_percents_filtered.values, labels=genus_percents_filtered.index, autopct='%1.1f%%', colors=colors_palette)
plt.title('Main Tree Genus in Finland')
plt.show()
```

```python
genus = ForestGenus[ForestGenus['COUNTRY'] == 'Norway']
genus_counts = genus['Genus'].value_counts()
genus_percents = genus_counts / genus_counts.sum()
genus_percents_filtered = genus_percents[genus_percents > 0.03]
colors_palette = genus_percents_filtered.index.map(colors_dict.get)
plt.pie(genus_percents_filtered.values, labels=genus_percents_filtered.index, autopct='%1.1f%%', colors=colors_palette)
plt.title('Main Tree Genus in Norway')
plt.show()
```

```python
genus = ForestGenus[ForestGenus['COUNTRY'] == 'Sweden']
genus_counts = genus['Genus'].value_counts()
genus_percents = genus_counts / genus_counts.sum()
genus_percents_filtered = genus_percents[genus_percents > 0.03]
colors_palette = genus_percents_filtered.index.map(colors_dict.get)
plt.pie(genus_percents_filtered.values, labels=genus_percents_filtered.index, autopct='%1.1f%%', colors=colors_palette)
plt.title('Main Tree Genus in Sweden')
plt.show()
```
