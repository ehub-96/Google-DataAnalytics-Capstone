### Cleaning Data

```python
genusiceland_weather
```

```python
genusiceland_weather.drop(['X', 'Y'], axis=1, inplace=True)
```

```python
genusiceland_weather
```

```python
genusiceland_weather.rename(columns={'Genus name': 'Genus'}, inplace=True)
```

```python
genusiceland_weather
```

```python
arcticgenuses = ['Abies','Alnus', 'Betula', 'Corylus','Juniperus', 'Larix', 'Picea', 'Pinus','Populus','Salix', 'Sorbus']
```

```python
genusiceland_weather = genusiceland_weather[genusiceland_weather["Genus"].isin(arcticgenuses)]
```

```python
genusiceland_weather
```

```python
speciesiceland_weather.drop(['X', 'Y'], axis=1, inplace=True)
```

```python
speciesiceland_weather
```

```python
speciesiceland_weather.rename(columns={'SPECIES NAME': 'SPECIES'}, inplace=True)
```

```python
speciesiceland_weather
```

```python
speciesiceland_weather.replace(-9999, 0, inplace=True)
```

```python
speciesiceland_weather
```

```python
speciesiceland_weather.SPECIES.unique()
```

```python
ArcticGenra = ['Alnus glutinosa', 'Alnus incana', 'Betula nana', 'Betula pendula', 'Betula pubescens', 'Corylus avellana', 'Juniperus communis', 'Larix sibirica', 'Picea abies', 'Pinus contorta', 'Pinus sylvestris', 'Populus tremula', 'Salix aurita', 'Salix caprea', 'Salix cinerea', 'Salix glauca', 'Salix hastata', 'Salix lanata', 'Salix lapponum', 'Salix myrsinifolia', 'Salix myrtilloides', 'Salix pentandra', 'Salix phylicifolia', 'Salix repens', 'Salix rosmarinifolia', 'Salix starkeana', 'Sorbus aucuparia', 'Sorbus intermedia']
```

```python
speciesiceland_weather[speciesiceland_weather["SPECIES"].isin(ArcticGenra)]
```

```python
speciesiceland_weather = speciesiceland_weather[speciesiceland_weather["SPECIES"].isin(ArcticGenra)]
```

```python
speciesiceland_weather
```
